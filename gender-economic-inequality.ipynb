# Gender Economic Inequality
Gender Economic Inequality referes to the disparties in income, wealth, and opportunities, between men and women. This can happen for many factors,such as including wage gaps, occupational segregation, and limited access to education and resources. Addressing these inequalities is crucial for achieving gender equality and fostering inclusive economic. 

The gender pay gap is a metric that shows the difference in pay between women and men. Gender pay gap measur of inequality and captures a concept that is broader than the concept of equal pay for equal work.

Discrimination in hiring practices can exist in the absence of pay gaps - for example, if women know they will be treated unfairly and hence choose not to participate in the labor market. Similarly, it is possible to observe large pay gaps in the absence of discrimination in hiring practices.

### count of occurrences of gender wage gap for different countries

```python
df['country'].value_counts().head(15).sort_values(ascending=True)
```

```python
# count of occurrences of gender wage gap for different countries
sns.set_style('darkgrid')
sns.color_palette("dark:#5A8", as_cmap=True)
plt.figure(figsize = (10,6))
country_counts = df['country'].value_counts().head(15).sort_values(ascending=True)

ax = country_counts.plot(kind='barh', color='#AC02A4', width = 0.8)
for i, count in enumerate(country_counts):
    ax.text(count + 0.5, i, str(count), va='center', fontweight = 'bold', color = '#14AC02')
    
plt.title("count of occurrences of gender wage gap for different countries", fontweight = 'bold', fontsize = 19, color = '#580C54')
plt.ylabel('country', fontweight = 'bold', fontsize = 15)
plt.xticks([])
plt.show()
```

**Count of occurrences of gender wage gap for different countries**
- Brazil, Argentina, Honduras, and Costa Rica have relatively higher counts, indicating either more available data or possibly more pronounced issues with gender wage gaps in these countries, leading to more research and reporting on the topic
- Several countries from South America, such as Uruguay, Panama, Dominican Republic, Peru, Paraguay, Bolivia, and Ecuador, are also represented in the list.
